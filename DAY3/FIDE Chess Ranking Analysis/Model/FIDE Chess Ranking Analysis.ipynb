# Visualisation of Data

```python
df.groupby('name ')['ELO '].mean().nlargest(20).plot.bar()
```

```python
df.groupby('name ')['games'].mean().nlargest(20).plot.bar()
```

```python
df.groupby('name ')['games'].mean().nsmallest(20).plot.bar()
```

# Plotting the Above data in Pie Chart

```python
plt.figure(figsize=(10, 10))
top = df['title'].value_counts().sort_values(ascending=False).head(13)
top.plot(kind='pie')
```

```python
plt.pie(df['federation '].value_counts(), labels=df['federation '].value_counts().index,counterclock=False,autopct='%.2f')
plt.title('Participants from each Federations');
```

```python
df.rename(columns = {'federation ':'federation'}, inplace = True)
```
